```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tabulate import tabulate
```

```python
path_a = "/content/a_Vout_at_K_V15Hz.csv"
a = pd.read_csv(path_a)
path_b = "/content/b_Pout_at_K_V15Hz.csv"
b = pd.read_csv(path_b)
path_c = "/content/c_Iout_at_K_V15Hz.csv"
c = pd.read_csv(path_c)
a, b, c
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Plot A
axes[0].plot(a['V2'], a['V_out'])
axes[0].set_xlabel('DC Vin [mV]')
axes[0].set_ylabel('Vout [V]')
axes[0].set_title('Ouput Voltage at 20mVAC, 5Hz and Varrying DC source')

# Plot B
axes[1].plot(b['V2'], b['P_out'])
axes[1].set_xlabel('DC Vin [mV]')
axes[1].set_ylabel('P_out [µW]')
axes[1].set_title('Power output at 20mVAC, 5Hz and Varrying DC source')

# Plot C
axes[2].plot(c['V2'], c['I_out'])
axes[2].set_xlabel('DC Vin [mV]')
axes[2].set_ylabel('Iout [µA]')
axes[2].set_title('Output current at 20mVAC, 5Hz and Varrying DC source')

plt.tight_layout()
fig.savefig('abc_plots')
plt.show();
```
